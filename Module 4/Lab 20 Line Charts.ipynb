### Task 1: Trends in Compensation Over Age Groups


##### 1. Line Chart of Median `ConvertedCompYearly` by Age Group


- Track how the median yearly compensation (ConvertedCompYearly) changes across different age groups.

- Use a line chart to visualize these trends.


```python
import sqlite3
conn = sqlite3.connect("survey-results-public.sqlite")
```

```python
## Write your code here

# Step 1: Fetch raw data
query = 'SELECT Age, ConvertedCompYearly FROM main'
df_raw = pd.read_sql_query(query, conn)

# Step 2: Group by Age and calculate median in Python
df_median = df_raw.groupby('Age')['ConvertedCompYearly'].median().reset_index()
df_median = df_median.sort_values(by='Age')

# Step 3: Plot the line chart
plt.figure(figsize=(10, 6))
plt.plot(df_median['Age'], df_median['ConvertedCompYearly'], marker='o', linestyle='-', color='teal')
plt.title('Median Yearly Compensation by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Median Compensation (USD)')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

##### 2. Line Chart of Median `ConvertedCompYearly` for Ages 25 to 45


For a closer look, plot a line chart focusing on the median compensation for respondents between ages 25 and 45.


```python
## Write your code here

# Step 1: Fetch raw data
query = 'SELECT Age, ConvertedCompYearly FROM main'
df_raw = pd.read_sql_query(query, conn)

# Step 2: Map age bands to numeric midpoints
age_map = {
    '18-24 years old': 21,
    '25-34 years old': 30,
    '35-44 years old': 40,
    '45-54 years old': 50,
    '55-64 years old': 60,
    '65 years or older': 70
}
df_raw = df_raw[df_raw['Age'].isin(age_map.keys())]
df_raw['AgeMid'] = df_raw['Age'].map(age_map)

# Step 3: Filter for ages 25–45
df_filtered = df_raw[df_raw['AgeMid'].between(25, 45)]

# Step 4: Group and calculate median
df_median = df_filtered.groupby('AgeMid')['ConvertedCompYearly'].median().reset_index()

# Step 5: Plot with annotations
plt.figure(figsize=(10, 6))
plt.plot(df_median['AgeMid'], df_median['ConvertedCompYearly'], marker='o', linestyle='-', color='blue')
plt.title('Median Yearly Compensation (Ages 25–45)')
plt.xlabel('Age (Midpoint of Band)')
plt.ylabel('Median Compensation (USD)')
plt.grid(True)
plt.xticks(df_median['AgeMid'])

# Annotate each point
for i, row in df_median.iterrows():
    plt.text(row['AgeMid'], row['ConvertedCompYearly'] + 1000,
             f"${int(row['ConvertedCompYearly']):,}", ha='center', fontsize=8, color='black')

plt.tight_layout()
plt.show()
```

### Task 2: Trends in Job Satisfaction by Experience Level



##### 1. Line Chart of Job Satisfaction (`JobSatPoints_6`) by Experience Level



- Use a column that approximates experience level to analyze how job satisfaction changes with experience.

- If needed, substitute an available experience-related column for `Experience`.


```python
## Write your code here

# Step 1: Fetch relevant data
query = 'SELECT YearsCodePro, JobSatPoints_6 FROM main'
df_raw = pd.read_sql_query(query, conn)

# Step 2: Clean experience column
def convert_experience(val):
    if val == 'Less than 1 year':
        return 0.5
    elif val == 'More than 50 years':
        return 51
    try:
        return float(val)
    except:
        return None

df_raw['YearsCodePro'] = df_raw['YearsCodePro'].apply(convert_experience)
df_clean = df_raw.dropna()

# Step 3: Group and calculate average job satisfaction
df_grouped = df_clean.groupby('YearsCodePro')['JobSatPoints_6'].mean().reset_index()

# Step 4: Plot the line chart
plt.figure(figsize=(10, 6))
plt.plot(df_grouped['YearsCodePro'], df_grouped['JobSatPoints_6'], marker='o', linestyle='-', color='mediumblue')
plt.title('Average Job Satisfaction by Experience Level')
plt.xlabel('Years of Professional Coding Experience')
plt.ylabel('Average Job Satisfaction (1–6 scale)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Task 3: Trends in Job Satisfaction and Compensation by Experience


##### 1.Line Chart of Median ConvertedCompYearly Over Experience Level

- This line chart will track how median compensation (`ConvertedCompYearly`) changes with increasing experience.

- Use a column such as `WorkExp` or another relevant experience-related column.


```python
## Write your code here

# Step 1: Fetch relevant data
query = 'SELECT YearsCodePro, ConvertedCompYearly FROM main'
df_raw = pd.read_sql_query(query, conn)


# Step 2: Clean experience column
def convert_experience(val):
    if val == 'Less than 1 year':
        return 0.5
    elif val == 'More than 50 years':
        return 51
    try:
        return float(val)
    except:
        return None

df_raw['YearsCodePro'] = df_raw['YearsCodePro'].apply(convert_experience)
df_clean = df_raw.dropna()

# Step 3: Group and calculate median compensation
df_grouped = df_clean.groupby('YearsCodePro')['ConvertedCompYearly'].median().reset_index()

# Step 4: Plot the line chart
plt.figure(figsize=(10, 6))
plt.plot(df_grouped['YearsCodePro'], df_grouped['ConvertedCompYearly'], marker='o', linestyle='-')
plt.title('Median Compensation by Experience Level')
plt.xlabel('Years of Professional Coding Experience')
plt.ylabel('Median Compensation (USD)')
plt.grid(True)
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

##### 2.Line Chart of Job Satisfaction (`JobSatPoints_6`) Across Experience Levels

- Create a line chart to explore trends in job satisfaction (`JobSatPoints_6`) based on experience level.

- This chart will provide insight into how satisfaction correlates with experience over time


```python
## Write your code here

# Step 1: Fetch relevant data
query = 'SELECT YearsCodePro, JobSatPoints_6 FROM main'
df_raw = pd.read_sql_query(query, conn)

# Step 2: Clean experience column
def convert_experience(val):
    if val == 'Less than 1 year':
        return 0.5
    elif val == 'More than 50 years':
        return 51
    try:
        return float(val)
    except:
        return None

df_raw['YearsCodePro'] = df_raw['YearsCodePro'].apply(convert_experience)
df_clean = df_raw.dropna()

# Step 3: Group and calculate average job satisfaction
df_grouped = df_clean.groupby('YearsCodePro')['JobSatPoints_6'].mean().reset_index()

# Step 4: Plot the line chart
plt.figure(figsize=(10, 6))
plt.plot(df_grouped['YearsCodePro'], df_grouped['JobSatPoints_6'], marker='o', linestyle='-', color='purple')
plt.title('Job Satisfaction Across Experience Levels')
plt.xlabel('Years of Professional Coding Experience')
plt.ylabel('Average Job Satisfaction (1–6 scale)')
plt.grid(True)
plt.xticks(rotation=45)

# Annotate each point
for i, row in df_grouped.iterrows():
    plt.text(row['YearsCodePro'], row['JobSatPoints_6'] + 0.5,
             f"{row['JobSatPoints_6']:.2f}", ha='center', fontsize=6)

plt.tight_layout()
plt.show()
```
