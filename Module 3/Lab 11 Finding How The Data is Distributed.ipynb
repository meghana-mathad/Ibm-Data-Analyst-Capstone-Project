### Step 3: Handle Missing Data


- Identify missing values in the dataset.

- Impute or remove missing values as necessary to ensure data completeness.



```python
## Write your code here
print(df.isnull().sum())
df_cleaned = df.dropna(subset=['Employment', 'JobSat', 'RemoteWork'])
df['Employment'] = df['Employment'].fillna(df['Employment'].mode()[0])
df['RemoteWork'] = df['RemoteWork'].fillna('Not specified')
df['ConvertedCompYearly'] = df['ConvertedCompYearly'].fillna(df['ConvertedCompYearly'].mean())
```

### Step 7: Analyze Remote Work Trends


- Visualize the distribution of RemoteWork by region using a grouped bar chart or heatmap.


```python
## Write your code here
# Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Group and count remote work responses by country
remote_by_region = df.groupby(['Country', 'RemoteWork']).size().unstack(fill_value=0)

# Step 3: Normalize to get proportions
remote_by_region_norm = remote_by_region.div(remote_by_region.sum(axis=1), axis=0)

# Step 4: Limit to top 10 countries by total responses (optional but helps spacing)
top_countries = remote_by_region.sum(axis=1).sort_values(ascending=False).head(10).index
remote_by_region_top = remote_by_region_norm.loc[top_countries]

# Step 5: Plot with better layout
plt.figure(figsize=(14,8))  # Wider figure for more space
remote_by_region_top.plot(kind='bar', stacked=False)

plt.title('Remote Work Distribution by Country')
plt.xlabel('Country')
plt.ylabel('Proportion of Respondents')

# Rotate and align x-axis labels
plt.xticks(rotation=45, ha='right')  # Rotate and right-align for readability

# Move legend outside the plot
plt.legend(title='Remote Work Type', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()


# Step 6: Heatmap
plt.figure(figsize=(12,8))
sns.heatmap(remote_by_region_top, annot=True, cmap='Blues', fmt=".2f")
plt.title('Remote Work Preferences by Region')
plt.xlabel('Remote Work Type')
plt.ylabel('Country')
plt.tight_layout()
plt.show()


```

### Step 8: Correlation between Job Satisfaction and Experience


- Analyze the correlation between overall job satisfaction (`JobSat`) and `YearsCodePro`.
  
- Calculate the Pearson or Spearman correlation coefficient.


```python
## Write your code here
import pandas as pd

# Convert YearsCodePro to numeric (if not already)
df['YearsCodePro'] = pd.to_numeric(df['YearsCodePro'], errors='coerce')

# Drop missing values
subset = df[['JobSat', 'YearsCodePro']].dropna()

# Calculate Pearson and Spearman correlation
pearson_corr = subset.corr(method='pearson').loc['JobSat', 'YearsCodePro']
spearman_corr = subset.corr(method='spearman').loc['JobSat', 'YearsCodePro']

print("Pearson correlation:", round(pearson_corr, 3))
print("Spearman correlation:", round(spearman_corr, 3))
```

### Step 9: Cross-tabulation Analysis (Employment vs. Education Level)


- Analyze the relationship between employment status (`Employment`) and education level (`EdLevel`).

- **Instruction**: Create a cross-tabulation using `pd.crosstab()` and visualize it with a stacked bar plot if possible.


```python
## Write your code here
# Step 1: Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Filter to top 10 most common employment categories to avoid overcrowding
top_employment = df['Employment'].value_counts().head(10).index
filtered_df = df[df['Employment'].isin(top_employment)]

# Step 3: Create cross-tabulation of Education Level vs Employment Status
cross_tab = pd.crosstab(filtered_df['EdLevel'], filtered_df['Employment'])

# Optional: Display raw counts
print("Cross-tabulation (counts):")
print(cross_tab)

# Step 4: Normalize to show proportions
cross_tab_norm = pd.crosstab(filtered_df['EdLevel'], filtered_df['Employment'], normalize='index')

# Optional: Display normalized table
print("\nCross-tabulation (proportions):")
print(cross_tab_norm)

# Step 5: Plot normalized cross-tab as a stacked bar chart
plt.figure(figsize=(20,10))  # Wider canvas to accommodate labels
cross_tab_norm.plot(kind='bar', stacked=True)

plt.title('Employment Status by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Proportion of Respondents')
plt.xticks(rotation=45, ha='right')  # Rotate and align x-axis labels
plt.legend(title='Employment Status', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

```
